```python
import pandas as pd
```

```python
cpv_path = 'demo2/sagemaker_data/files_meta_cpv.csv'
cpv_df = pd.read_csv(cpv_path)
mrna_path = 'demo2/sagemaker_data/files_meta_mRNA.csv'
mrna_df = pd.read_csv(mrna_path)
mirna_path = 'demo2/sagemaker_data/files_meta_miRNA.csv'
mirna_df = pd.read_csv(mirna_path)
```

```python
cpv_samples = cpv_df['cases.0.samples.0.sample_id']
mrna_samples = mrna_df['cases.0.samples.0.sample_id']
mirna_samples = mirna_df['cases.0.samples.0.sample_id']
all_samples = [cpv_samples,mrna_samples,mirna_samples]
```

```python
len(all_samples)
```

```python
common = set.intersection(*map(set,all_samples))
```

```python
common_list = list(common)
```

```python
len(common_list)
```

```python
# filter cpv 
```

```python
cpv_df=cpv_df[cpv_df['cases.0.samples.0.sample_id'].isin(common_list)]
```

```python
cpv_df.columns
```

```python
data_groups = cpv_df.groupby("cases.0.samples.0.sample_id")
size = data_groups.size().reset_index()
```

```python
size.columns
```

```python
len(size)
```

```python
size = size[size[0] > 1]
```

```python
duplicate_list = size['cases.0.samples.0.sample_id'].tolist()
```

```python
len(duplicate_list)
```

```python
df_out = pd.DataFrame({'sample_id':common_list})
# print (df_out)
```

```python
outputfile = 'common_sample_ids.csv'
df_out.to_csv(outputfile,index=False)
```

```python
cpv_path = 'CPV/cpv_final_matrix.csv'
cpv_df = pd.read_csv(cpv_path)
```

```python
cpv_df=cpv_df[cpv_df['sample_id'].isin(common_list)]
```

```python
len(cpv_df['sample_id']) # sample_id in common_list (10354)
```

```python
cpv_df.columns
```

```python
target_avg_col = cpv_df.columns[1:-4]
```

```python
cpv_df[cpv_df['sample_id'].isin(duplicate_list)]
```

```python
cpv_df_avg=cpv_df.groupby('sample_id').mean().reset_index()
```

```python
cpv_df_avg.columns
```

```python
len(cpv_df_avg)
```

```python
label_df=cpv_df[['sample_id','sample_type', 'disease_type', 'primary_diagnosis', 'case_id']]
```

```python
label_df=label_df.drop_duplicates(keep='first')
```

```python
len(label_df)
```

```python
label_df.columns
```

```python
cpv_final=pd.merge(cpv_df_avg, label_df, on='sample_id', how='inner')
```
